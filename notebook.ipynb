### **Cleaning Data**

```python
del hr_df
```

```python
drop_col = ['instant', 'windspeed']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

```python
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

```python
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

```python
day_df['dateday'] = pd.to_datetime(day_df.dateday)

day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')

day_df.head()
```

## **Visualization & Explanatory Analysis**

### **1. Bagaimana cuaca dapat memengaruhi jumlah penggunaan sepeda?**

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='weather_cond',
    y='count',
    data=day_df)

plt.title('Jumlah Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

Visualisasi data menggunakan boxplot menunjukkan adanya hubungan yang signifikan antara kondisi cuaca dan jumlah penyewaan sepeda. Cuaca cerah/setengah mendung menjadi kondisi yang paling banyak diminati penyewa, diikuti oleh cuaca berkabut/mendung, sedangkan cuaca bersalju/hujan hanya diminati sedikit penyewa.

### **2. Bagaimana pola penggunaan sepeda pada hari kerja biasa, hari libur, dan hari kerja khusus?**

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15,10))

# Berdasarkan workingday
sns.barplot(
    x='workingday',
    y='count',
    data=day_df,
    ax=axes[0])
axes[0].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Jumlah Pengguna Sepeda')

# Berdasarkan holiday
sns.barplot(
  x='holiday',
  y='count',
  data=day_df,
  ax=axes[1])
axes[1].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Libur')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Jumlah Pengguna Sepeda')

# Berdasarkan weekday
sns.barplot(
  x='weekday',
  y='count',
  data=day_df,
  ax=axes[2])
axes[2].set_title('Jumlah Pengguna Sepeda berdasarkan Hari dalam Seminggu')
axes[2].set_xlabel('Hari dalam Seminggu')
axes[2].set_ylabel('Jumlah Pengguna Sepeda')

plt.tight_layout()
plt.show()
```

Barplot menunjukkan beberapa pola dalam penggunaan sepeda. Pada hari kerja (Senin - Jumat), jumlah penyewa lebih tinggi dibandingkan akhir pekan (Sabtu dan Minggu). Penyewaan sepeda juga lebih banyak pada hari biasa dibandingkan hari libur nasional. Hari Jumat tercatat sebagai hari dengan jumlah penyewa terbanyak, sementara hari Minggu menjadi hari dengan jumlah penyewa paling sedikit.

### **3. Apakah terdapat perbedaan tren penggunaan sepeda antara tahun 2011 dan 2012, dan mana yang lebih tinggi?**

```python
day_df['month'] = pd.Categorical(day_df['month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

monthly_counts = day_df.groupby(by=["month","year"]).agg({
    "count": "sum"
}).reset_index()

sns.lineplot(
    data=monthly_counts,
    x="month",
    y="count",
    hue="year",
    palette="rocket",
    marker="o")

plt.title("Jumlah total sepeda yang disewakan berdasarkan Bulan dan tahun")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

Dari lineplot terlihat bahwa penggunaan sepeda di tahun 2012 (angka 1) lebih tinggi sepanjang tahun dibandingkan 2011. Puncak penyewaan sepeda di tahun 2012 terjadi di bulan September, sedangkan Januari memiliki penyewaan terendah. Pada tahun 2011, puncak penyewaan terjadi di bulan Juni dan jumlah penyewaan terendah ada di bulan Januari.

### **4. Seberapa besar pengaruh variabel suhu (temp dan atemp) serta kelembapan (humidity) terhadap jumlah pengguna sepeda, baik untuk pengguna Casual maupun Registered?**

```python
plt.figure(figsize=(14, 6))

# Scatter plot untuk 'temp' vs 'count'
plt.subplot(1, 3, 1)
sns.scatterplot(
    x='temp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Temperature vs Count')

# Scatter plot untuk 'atemp' vs 'count'
plt.subplot(1, 3, 2)
sns.scatterplot(
    x='atemp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Feels Like Temperature vs Count')

# Scatter plot untuk 'hum' vs 'count'
plt.subplot(1, 3, 3)
sns.scatterplot(
    x='hum',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Humidity vs Count')
```

Analisis menggunakan scatter plot menunjukkan hubungan positif antara suhu (temp dan atemp) dengan jumlah pengguna sepeda, di mana peningkatan suhu cenderung diikuti oleh peningkatan jumlah penyewa. Sementara itu, hubungan antara kelembapan (hum) dan jumlah pengguna menunjukkan korelasi sedikit negatif, tetapi tidak terlalu signifikan.

### **5. Apakah musim memiliki dampak yang berbeda terhadap jumlah penyewa sepeda pada kategori pelanggan Casual dan Registered?**

```python
# Mengelompokkan data berdasarkan musim dan menghitung jumlah penggunaan terdaftar dan tidak terdaftar
seasonal_usage = day_df.groupby('season')[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10, 6))

# Membuat
plt.bar(
    seasonal_usage['season'],
    seasonal_usage['registered'],
    label='Registered',
    color='tab:blue'
)

plt.bar(
    seasonal_usage['season'],
    seasonal_usage['casual'],
    label='Casual',
    color='tab:orange'
)

plt.xlabel(None)
plt.ylabel(None)
plt.title('Jumlah penyewaan sepeda berdasarkan musim')
plt.legend()
plt.show()
```

Dari visualisasi barplot, terlihat bahwa musim gugur (Fall) adalah musim yang paling disukai oleh pengguna sepeda baik untuk kategori Casual maupun Registered, diikuti oleh musim panas (Summer), musim dingin (Winter), dan terakhir musim semi (Spring).
